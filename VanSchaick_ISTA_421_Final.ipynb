```python
# importing packages and reading csv file
import numpy as num
import pandas as pan
import matplotlib.pyplot as plot

data = pan.read_csv("Console Generation data.csv")
```

```python
# removing commas and $ from each applicable column

data[' 2022 Price (Adjusted for inflation) '] = data[' 2022 Price (Adjusted for inflation) '].astype(str)
data[' 2022 Price (Adjusted for inflation) '] = data[' 2022 Price (Adjusted for inflation) '].str.replace(r'[^0-9.]','', regex=True).astype(float)
data[' 2022 Price (Adjusted for inflation) '] = data[' 2022 Price (Adjusted for inflation) '].astype(float)
data[' Total Systems Sold '] = data[' Total Systems Sold '].astype(str)
data[' Total Systems Sold '] = data[' Total Systems Sold '].str.replace(r'[^0-9.]','', regex=True).astype(float)
data[' Total Systems Sold '] = data[' Total Systems Sold '].astype(float)
data[' Original Price '] = data[' Original Price '].astype(str)
data[' Original Price '] = data[' Original Price '].str.replace(r'[^0-9.]','', regex=True).astype(float)
data[' Original Price '] = data[' Original Price '].astype(float)
```

```python
# adding new column for publishers
data['Publisher'] = ['Magnavox','Atari','Coleco','Fairchild Camera and Instrument','Atari','Mattel','Magnavox','Coleco','Nintendo','Sega','Atari','NEC','Sega','SNK','Nintendo','Sega','Nintendo','Sega','Sony','Nintendo','Nintendo','Sega','Sony','Nintendo','Nintendo','Microsoft','Microsoft','Sony','Nintendo','Sony','Nintendo','Nintendo','Sony','Microsoft','Nintendo','Sony','Nintendo','Microsoft','Microsoft','Sony','Sony']

data
```

```python


# Performing Lasso below
# I really tried to figure out how to do multi linear regression from scratch, and I just kept hitting a brick wall
# I used packages for multi linear regression, Lasso, and R^2
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# x and y variables set
x = data[['Year of release',' Original Price ',' 2022 Price (Adjusted for inflation) ']]
y = data[' Total Systems Sold ']
'''
# splitting data as close to 75/25 as possible (31 train, 10 test)
shuffled_data = data.sample(frac=1, random_state=num.random.seed(42))
train_split = shuffled_data[:31]
test_split = shuffled_data[31:]

# standardizing x and y so all predictors are affected equally
x_mean = x.mean(axis = 0)
x_std = x.std(axis = 0)
standard_x = (x - x_mean / x_std)

y_mean = y.mean()
y_centered = y - y_mean

# multi linear regression equation
# finding the coefficients

B0 = (1 * 1)
'''
#creating lasso lambda grid
'''lassolambda_grid = LassoCV(alphas=np.linspace(10**-2, 10**10, 100), cv=5)'''

# using packages
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=100)

model = LinearRegression()
model.fit(x_train, y_train)
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)

test_r_squared = model.score(x_test, y_test)
train_r_squared = model.score(x_train, y_train)
print('Test R^2:',test_r_squared)
print('Train R^2:',train_r_squared)
```

```python
# applying lasso, hopefully it can fix these terrible R scores

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso

# scaling predictors so none are unfairly affected
scale = StandardScaler()
scaled_x_train = scale.fit_transform(x_train)
scaled_x_test = scale.transform(x_test)

#cross validating to find best lambda value
lasso_cv = LassoCV(cv=5, random_state=100)
lasso_cv.fit(scaled_x_train, y_train)
best_lambda = lasso_cv.alpha_
print('Best Lambda Value:', best_lambda)

#fit the lambda value and then print the Lasso intercept and coefficients
lasso = Lasso(alpha=best_lambda)
lasso.fit(scaled_x_train, y_train)
print('Lasso Intercept:',lasso.intercept_)
print('Lasso Coefficients:', lasso.coef_)

#checking R^2 after lasso and printing both test and training r squared values
train_r2 = lasso.score(scaled_x_train, y_train)
test_r2 = lasso.score(scaled_x_test, y_test)

print("Lasso Train R^2:", train_r2)
print("Lasso Test R^2:", test_r2)
```

```python
# the zero here represents the Original Price column, meaning that it has no significance in predicting Total Console Sales
# 2 predictors to use are Year of release and 2022 Price
# R^2 values are still bad, meaning this model is not reliable at predicting total console sales at all
# Improved with Lasso, but still not good

#creating a scatterplot of the test and predicted y values to show accuracy of model (based off of those
# lasso r^2 values it's still not accurate, however it is more accurate than before lasso)
# Scatterplot for visualization
import matplotlib.pyplot as plot
y_pred = lasso.predict(scaled_x_test)

plot.scatter(y_test, y_pred)
plot.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()])

plot.xlabel('Actual Total Systems Sold')
plot.ylabel('Predicted Total Systems Sold')
plot.title('Lasso: Actual VS Predicted')
plot.show
```
